### <u>Sobre a base de dados</u>

* Os dados referem-se a posts publicados durante o ano de 2014 na página do Facebook de uma renomada marca de cosméticos.


* Este conjunto de dados contém 500 das 790 linhas e parte dos recursos analisados pelo artigo de MORO et al. (2016). O restante dos dados foram omitidos por questões de confidencialidade. 


* Inclui 7 indicadores conhecidos antes da postagem publicada e 12 indicadores para avaliação do impacto após a publicação (ver Tabelas 2 e 3 de Moro et al., 2016 - referência completa no 'Pedido de Citação').

<a id='dicionario'></a>

### <u>Dicionário da base de dados</u>

<br>

Este dataset possui `500 amostras` com `19 atributos`:

<br>

**1.** `Page total likes` - Quantidade de pessoas que curtiram a página da empresa.


**2.** `Type` - Tipo de conteúdo (link, foto, status, vídeo).


**3.** `Category` - Caracterização do conteúdo manual: ação (ofertas especiais e concursos), produto (propaganda direta, conteúdo explícito da marca) e inspiração (conteúdo não explícito relacionado à marca).


**4.** `Post Month` - Mês em que a postagem foi publicada (janeiro, fevereiro, março, ... , dezembro).


**5.** `Post Weekday` - Dia da semana em que a postagem foi publicada (domingo, segunda, ... , sábado).


**6.** `Post Hour` - Hora em que a postagem foi publicada (0, 1, 2, 3, 4, ... , 23).


**7.** `Paid` - Se a empresa pagou ao Facebook para publicidade (sim ou não)


**8.** `Lifetime Post Total Reach` - O número de pessoas que viram uma postagem na página (usuários únicos).


**9.** `Lifetime Post Total Impressions` - Impressões são o número de vezes que uma postagem de uma página é exibida, independentemente de a postagem ser clicada ou não. As pessoas podem ver várias impressões da mesma postagem. Por exemplo, alguém pode ver uma atualização de página no Feed de notícias uma vez e, em seguida, uma segunda vez se um amigo a compartilhar. 


**10.** `Lifetime Engaged Users` -  O número de pessoas que clicaram em qualquer lugar em uma postagem (usuários únicos).


**11.** `Lifetime Post Consumers` - O número de pessoas que clicaram em qualquer lugar em uma postagem.


**12.** `Lifetime Post Consumptions` - O número de cliques em qualquer lugar em uma postagem.


**13.** `Lifetime Post Impressions by people who have liked your Page` - Número total de impressões apenas de pessoas que curtiram uma página.

**14.** `Lifetime Post reach by people who like your Page` - O número de pessoas que viram uma postagem na página prque <u>gostaram</u> dela (usuários únicos).

**15.** `Lifetime People who have liked your Page and engaged with your post` - O número de pessoas que curtiram uma página e clicaram em qualquer lugar em uma postagem (usuários únicos)

**16.** `comment` - Número de comentários sobre a publicação.

**17.** `like` - Número de "curtidas" na publicação.

**18.** `share` - Número de vezes que a publicação foi compartilhada. 

**19.** `Total Interactions` - A soma de "curtidas", "comentários" e "compartilhamentos" da postagem.

## Gráficos com Matplotlib

### <u>Histogramas</u>

````python
fig = plt.figure(figsize=(20, 5)) # o tamanho da imagem está ok?
plt.hist(dados_midia.loc[:, 'comment'], bins=30)
plt.show()
````

### Gera múltiplos histogramas

````python
fig, axs = plt.subplots(1, 5, sharey=True, tight_layout=True, figsize=(15, 5))

axs[0].hist(dados_midia.loc[:, 'comment'], bins=30)
axs[1].hist(dados_midia.loc[:, 'like'], bins=30)
axs[2].hist(dados_midia.loc[:, 'share'], bins=30)
axs[3].hist(dados_midia.loc[:, 'Total Interactions'], bins=30)
axs[4].hist(dados_midia.loc[:, 'Lifetime Post Total Reach'], bins=30)

plt.show()
````

### Formata título e nome dos eixos

````python
fig = plt.figure(figsize = (20, 5))
plt.hist(dados_midia.loc[:, 'comment'], bins=30, color='green') 
plt.title('Histograma da Variável Comentários') # o tamanho da fonte está ok?
plt.xlabel('Quantidade de comentários')
plt.ylabel('Frequência')
plt.show()
````

### <u>Gráfico de barras</u>

````python
fig = plt.figure(figsize = (20, 5))

plt.bar(dados_midia.loc[:, 'comment'], dados_midia.loc[:, 'Lifetime Post Total Reach']) 

plt.xlabel('Quantidade de comentários')
plt.ylabel('Número de pessoas que viram uma página')
plt.title('Alcance do post & #Comentários')
plt.show()
````

### Formatando o gráfico

````python
fig = plt.figure(figsize = (15, 5))
plt.bar(x=dados_midia.loc[:, 'comment'], height=dados_midia.loc[:, 'Lifetime Post Total Reach'],
        width=2.5, color='orange')
plt.xlabel('Quantidade de comentários', size='14')
plt.ylabel('Número de pessoas que viram uma na página', size='14')
plt.title('Alcance do post vs Qtde. de comentários', size='18')
plt.show()
````

### <u>Gráfico de dispersão</u>

````python
fig = plt.figure(figsize = (20, 5))

plt.scatter(x=dados_midia.loc[:, 'comment'], y=dados_midia.loc[:, 'Lifetime Post Total Reach'],
        s=100, linewidths=1, color='orange', edgecolors='gray')
plt.xlabel('Quantidade de comentários', size='14')
plt.ylabel('Número de pessoas que viram uma página', size='14')
plt.title('Alcance do post vs Qtde. de comentários', size='18')
plt.show()
````

````python
fig = plt.figure(figsize = (20, 5))

plt.plot(dados_midia.groupby(by='Type').count().index,
         dados_midia.groupby(by='Type').count().loc[:, 'comment'],
         color='green', marker='o', linestyle='dashed',
         linewidth=2, markersize=12)

plt.xlabel('Tipo de post', size='14')
plt.ylabel('Número de comentários', size='14')
plt.title('Alcance do post vs Qtde. de comentários', size='18')
plt.show()
````
